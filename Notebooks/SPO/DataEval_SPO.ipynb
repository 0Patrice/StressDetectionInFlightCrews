```python
sc1_lens = []
sc2_lens = []
sc3_lens = []
sc4_lens = []
sc5_lens = []
```

```python
# Find Runtime
folder_path = 'Data/SPO/MultiRecordFiles/'

for root, subdirs, files in os.walk(folder_path, topdown=True):
    for file in files:
        
        if file.endswith('Sc1MRF_mV.csv'):
            print(file)
            file_path = os.path.join(root, file)
            df_t = pd.read_csv(file_path, delimiter=';', usecols=columns)
            print(len(df_t))
            sc1_lens.extend([len(df_t)])

        elif file.endswith('Sc2MRF_mV.csv'):
            print(file)
            file_path = os.path.join(root, file)
            df_t = pd.read_csv(file_path, delimiter=';', usecols=columns)
            print(len(df_t))
            sc2_lens.extend([len(df_t)])
            
        elif file.endswith('Sc3MRF_mV.csv'):
            print(file)
            file_path = os.path.join(root, file)
            df_t = pd.read_csv(file_path, delimiter=';', usecols=columns)
            print(len(df_t))
            sc3_lens.extend([len(df_t)])
            
        elif file.endswith('Sc4MRF_mV.csv'):
            print(file)
            file_path = os.path.join(root, file)
            df_t = pd.read_csv(file_path, delimiter=';', usecols=columns)
            print(len(df_t))
            sc4_lens.extend([len(df_t)])
            
        elif file.endswith('Sc5MRF_mV.csv'):
            print(file)
            file_path = os.path.join(root, file)
            df_t = pd.read_csv(file_path, delimiter=';', usecols=columns)
            print(len(df_t))
            sc5_lens.extend([len(df_t)])
            
        else:
            print(file)
```
