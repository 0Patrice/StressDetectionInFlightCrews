```python
import os
import numpy as np
import pandas as pd
from wfdb import processing
%config InlineBackend.figure_formats = ['svg']
import mplcursors
import colorcet as cc

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns 
```

```python
folder_path = '../../Data/SPO/MultiRecordFiles'

columns = ['Time', 'Value', 'Stress']



hist_sc1_array = []
hist_sc2_array = []
hist_sc3_array = []
hist_sc4_array = []
hist_sc5_array = []

hr_sc1_array = []
hr_sc2_array = []
hr_sc3_array = []
hr_sc4_array = []
hr_sc5_array = []

sampling_rate = 300

sl_list = []
hr_list = []
vp_list = []
scenario_list = []
file_list = []
```

```python
def saveFigureInPDF(hr_list, sl_list, file_name):
    plt.figure()
    plt.rcParams["figure.figsize"] = (20,10)
    #palette = sns.color_palette(cc.glasbey_hv, n_colors=105)

    sns.scatterplot(x=sl_list, y=hr_list, s=10)
    plt.legend(bbox_to_anchor=(0.95, 1), loc='upper left', borderaxespad=0)
    legend.remove()
    
    plt.title(file_name)
    plt.xlim([-1,11])

    plt.xlabel('Stress Level')
    plt.ylabel('Heart Rate')
    plt.subplots_adjust(
        left=0.035,
        bottom=0.05,
        right=0.975,
        top=0.975,
        wspace=0.05,
        hspace=0)
    file_name = file_name.removesuffix('MRF_mV.csv')
    plt.savefig(f"PDF/{file_name}_HRSL.pdf", format="pdf")
    
```

```python
for root, subdirs, files in os.walk(folder_path, topdown=True):
    for file in files:
        
        if file.endswith('Sc1MRF_mV.csv'):
            print(file)
            #print('Scenario 1')
            file_path = os.path.join(root, file)
            
            df = pd.read_csv(file_path, delimiter=';', usecols=columns)
            
            print(f'Shape: {df.shape}')
            
            
            xqrs = processing.XQRS(sig=df['Value'], fs=sampling_rate)
            
            xqrs.detect()

            # XQRS inds in samples -> to seconds divide by sampling rate
            # From seconds to ms multipy by 1000
            rr = np.diff(xqrs.qrs_inds)*1000 / sampling_rate

            # From ms to min divide rr by 60000
            heart_rate_r2 = 60 * 1000 / rr
            
            sl_temp = df.loc[xqrs.qrs_inds, 'Stress'].tolist()
            sl_temp.pop(0)
                        
            file_names = [file for i in range(len(sl_temp))]
            
            scenario_names = ['Scenario1' for i in range(len(sl_temp))]
            
            t_vp_name = file.removesuffix('Sc1MRF_mV.csv')
            t_vp_names = [t_vp_name for i in range(len(sl_temp))]


            file_list.extend(file_names)
            
            hist_sc1_array.extend(df['Stress'].values)
            
            sl_list.extend(sl_temp)
            hr_list.extend(heart_rate_r2)
            
            vp_list.extend(t_vp_names)
            
            scenario_list.extend(scenario_names)
            
            hr_sc1_array.extend(heart_rate_r2)
            
            #saveFigureInPDF(heart_rate_r2, sl_temp, file)
            
        elif file.endswith('Sc2MRF_mV.csv'):
            print(file)
            #print('Scenario 2')
            
            file_path = os.path.join(root, file)
            
            df = pd.read_csv(file_path, delimiter=';', usecols=columns)
            
            hist_sc2_array.extend(df['Stress'].values)
                        
            xqrs = processing.XQRS(sig=df['Value'], fs=sampling_rate)
            
            xqrs.detect()

            # XQRS inds in samples -> to seconds divide by sampling rate
            # From seconds to ms multipy by 1000
            rr = np.diff(xqrs.qrs_inds)*1000 / sampling_rate

            # From ms to min divide rr by 60000
            heart_rate_r2 = 60 * 1000 / rr
            
            sl_temp = df.loc[xqrs.qrs_inds, 'Stress'].tolist()
            sl_temp.pop(0)
            
            sl_list.extend(sl_temp)
            hr_list.extend(heart_rate_r2)
            
            t_vp_name = file.removesuffix('Sc2MRF_mV.csv')
            t_vp_names = [t_vp_name for i in range(len(sl_temp))]
            
            file_names = [file for i in range(len(sl_temp))]
            scenario_names = ['Scenario2' for i in range(len(sl_temp))]
            
            file_list.extend(file_names)
            hr_sc2_array.extend(heart_rate_r2)
            vp_list.extend(t_vp_names)
            scenario_list.extend(scenario_names)
            
            #saveFigureInPDF(heart_rate_r2, sl_temp, file)
            
        elif file.endswith('Sc3MRF_mV.csv'):
            print(file)
            #print('Scenario 3')
            
            file_path = os.path.join(root, file)
            
            df = pd.read_csv(file_path, delimiter=';', usecols=columns)
            
            hist_sc3_array.extend(df['Stress'].values)
                        
            xqrs = processing.XQRS(sig=df['Value'], fs=sampling_rate)
            
            xqrs.detect()

            # XQRS inds in samples -> to seconds divide by sampling rate
            # From seconds to ms multipy by 1000
            rr = np.diff(xqrs.qrs_inds)*1000 / sampling_rate

            # From ms to min divide rr by 60000
            heart_rate_r2 = 60 * 1000 / rr
            
            sl_temp = df.loc[xqrs.qrs_inds, 'Stress'].tolist()
            sl_temp.pop(0)
            
            sl_list.extend(sl_temp)
            hr_list.extend(heart_rate_r2)
            
            t_vp_name = file.removesuffix('Sc3MRF_mV.csv')
            t_vp_names = [t_vp_name for i in range(len(sl_temp))]
            
            file_names = [file for i in range(len(sl_temp))]
            scenario_names = ['Scenario3' for i in range(len(sl_temp))]
            
            file_list.extend(file_names)
            hr_sc3_array.extend(heart_rate_r2)
            
            vp_list.extend(t_vp_names)
            scenario_list.extend(scenario_names)
            
            #saveFigureInPDF(heart_rate_r2, sl_temp, file)
            
        elif file.endswith('Sc4MRF_mV.csv'):
            print(file)
            #print('Scenario 4')
            
            file_path = os.path.join(root, file)
            
            df = pd.read_csv(file_path, delimiter=';', usecols=columns)
            
            hist_sc4_array.extend(df['Stress'].values)
                        
            xqrs = processing.XQRS(sig=df['Value'], fs=sampling_rate)
            
            xqrs.detect()

            # XQRS inds in samples -> to seconds divide by sampling rate
            # From seconds to ms multipy by 1000
            rr = np.diff(xqrs.qrs_inds)*1000 / sampling_rate

            # From ms to min divide rr by 60000
            heart_rate_r2 = 60 * 1000 / rr
            
            sl_temp = df.loc[xqrs.qrs_inds, 'Stress'].tolist()
            sl_temp.pop(0)
            
            sl_list.extend(sl_temp)
            hr_list.extend(heart_rate_r2)

            t_vp_name = file.removesuffix('Sc4MRF_mV.csv')
            t_vp_names = [t_vp_name for i in range(len(sl_temp))]
            
            file_names = [file for i in range(len(sl_temp))]
            scenario_names = ['Scenario4' for i in range(len(sl_temp))]
            
            file_list.extend(file_names)
            hr_sc4_array.extend(heart_rate_r2)
            
            vp_list.extend(t_vp_names)
            scenario_list.extend(scenario_names)
            
            #saveFigureInPDF(heart_rate_r2, sl_temp, file)
            
        elif file.endswith('Sc5MRF_mV.csv'):
            print(file)
            #print('Scenario 5')
            
            file_path = os.path.join(root, file)
            
            df = pd.read_csv(file_path, delimiter=';', usecols=columns)
            
            hist_sc5_array.extend(df['Stress'].values)
                        
            xqrs = processing.XQRS(sig=df['Value'], fs=sampling_rate)
            
            xqrs.detect()

            # XQRS inds in samples -> to seconds divide by sampling rate
            # From seconds to ms multipy by 1000
            rr = np.diff(xqrs.qrs_inds)*1000 / sampling_rate

            # From ms to min divide rr by 60000
            heart_rate_r2 = 60 * 1000 / rr
            
            sl_temp = df.loc[xqrs.qrs_inds, 'Stress'].tolist()
            sl_temp.pop(0)
            
            sl_list.extend(sl_temp)
            hr_list.extend(heart_rate_r2)
            

            t_vp_name = file.removesuffix('Sc5MRF_mV.csv')
            t_vp_names = [t_vp_name for i in range(len(sl_temp))]
            
            file_names = [file for i in range(len(sl_temp))]
            scenario_names = ['Scenario5' for i in range(len(sl_temp))]
            
            file_list.extend(file_names)
            hr_sc5_array.extend(heart_rate_r2)
            
            vp_list.extend(t_vp_names)
            scenario_list.extend(scenario_names)
            
            #saveFigureInPDF(heart_rate_r2, sl_temp, file)
            
        else:
            print(file)
            #print('Different File')
```
