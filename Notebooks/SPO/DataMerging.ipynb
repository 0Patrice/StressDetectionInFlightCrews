```python
import os
import sys
import numpy as np
from scipy import signal, interpolate
import csv

from xml.dom import minidom

from filereader import gen4file
```

```python

fs_gal = 300
fs_global = 1000

newline_char = '\n'
delimiter_char = ';'


data_folder = "/Volumes/LeMonde/DataEval/RawData"

outermarker = 1158.24 #3800ft in meter
middlemarker = 762 #@2500ft in meter
innermarker = 304.8 #@1000 in meter

edds_elevation = 359.968
eddf_elevation = 110.95 #### To be checked !!!!!! 
```

```python

for root, subdirectories, files in os.walk(data_folder, topdown=False):
    
   #for subdirectory in subdirectories:
        
        #print(os.path.join(root, subdirectory))
        
        #print(subdirectory)
    
    split1 = os.path.split(root)
    split2 = os.path.split(split1[0])
    print(split1[1])
    print(split2[1])
    print(split1[0])
    print(split2[0])
    
    csvfile = ""
    gen4filename = ""
    recafile = ""
    svgfile = ""
    
    
    # Get File from the Root Directory and store their names
    for file in files:
        print("---------------- New File ----------------")
        print(file)
        
        if file.endswith(".csv") and not file.startswith('.'):
            print("---------------- CSV ----------------")
            csvfile = file
            print(file)
            
        if file.endswith(".gen4") and not file.startswith('.'):
            print("---------------- GEN 4 ----------------")
            gen4filename = file
            print(file)
        
        if file.endswith(".reca") and not file.startswith('.'):
            print("---------------- REC A ----------------")
            recafile = file
            print(file)
            
        if file.endswith(".svg") and not file.startswith('.'):
            print("---------------- SVG ----------------")
            svgfile = file
            print(file)
    
    
    print('---- Eval ----')
    
    # Indecies for the export File
    print(split1[1][-1])
    print(split2[1][-2]+split2[1][-1])
    
    print(csvfile)
    print(gen4filename)
    print(recafile)
    print(svgfile)
    
    if recafile != "":
        #Stress Profile:
        svgfilepath = os.path.join(root, svgfile)
        print(svgfilepath)
        pst_file = minidom.parse(svgfilepath)

        svgfileheader = pst_file.getElementsByTagName('svg')

        viewbox = svgfileheader[0].attributes['viewBox'].value

        viewboxElems = viewbox.split(' ')

        y_max = float(viewboxElems[3])
        x_max = float(viewboxElems[2])

        print(f'Y Max: {y_max}')
        print(f'X Max: {x_max}')
        itemlist = pst_file.getElementsByTagName('polyline')

        pointList = itemlist[0].attributes['points'].value

        pointarry = pointList.split(' ')
        pointsarray = []
        for points in pointarry:
            if points:
                sepPoints = points.split(',')
                pointsarray.append(list(map(float, sepPoints)))

        pointsarray = np.array(pointsarray)


        if split1[1] == "Scenario5":
            print("!!!!!!! Scenario5 !!!!!!!")
            
            

        else:

            linelist = pst_file.getElementsByTagName('line')

            stressMarker = []

            for line in linelist:
                value = line.attributes['x1'].value
                stressMarker.append(float(value))
                print(line.attributes['x1'].value)


            print(stressMarker)

            # Sim File
            simfilepath = filePath = os.path.join(root, recafile)

            myFile = np.genfromtxt(simfilepath, delimiter=',', skip_header=128)

            # Load System Time
            sysTime = myFile[:, 0]

            # normalize System Time
            sysTime = np.array(sysTime)-sysTime[0]
            
            # Get Altitude from Z GPS Pos
            altitudeGPSz = myFile[:, 109]

            # outer
            outer_idx = (np.abs((-altitudeGPSz) - outermarker - edds_elevation)).argmin()
            print(outer_idx)

            om_time = round(sysTime[outer_idx],2)

            print(f"OM at {om_time} s")

            # middle
            middle_idx = (np.abs((-altitudeGPSz) - middlemarker - edds_elevation)).argmin()
            print(middle_idx)

            mm_time = round(sysTime[middle_idx],2)
            print(f"MM at {mm_time} s")

            # inner
            inner_idx = (np.abs((-altitudeGPSz) - innermarker - edds_elevation)).argmin()
            print(inner_idx)
            im_time = round(sysTime[inner_idx],2)
            print(f"IM at {im_time} s")

            for points in pointsarray:
                if points[0] < stressMarker[2]:

                    points[1] = ((points[1]/ -y_max + 1)*10)

                    points[0] = (points[0]/stressMarker[2])*om_time

                    print(f"S - OM {points}")

                elif points[0] > stressMarker[2] and points[0] < stressMarker[1]:

                    points[1] = ((points[1]/ -y_max + 1)*10)
                    points[0] = ((points[0]-stressMarker[2])/(stressMarker[1]-stressMarker[2]))*(mm_time-om_time)+om_time

                    print(f"OM - MM {points}")
                elif points[0] > stressMarker[1] and points[0] < stressMarker[0]:

                    points[1] = ((points[1]/ -y_max + 1)*10)
                    points[0] = ((points[0]-stressMarker[1])/(stressMarker[0]-stressMarker[1]))*(im_time-mm_time)+mm_time

                    print(f"MM - IM {points}")

                elif points[0] > stressMarker[0]:

                    points[1] = ((points[1]/ -y_max + 1)*10)
                    points[0] = ((points[0]-stressMarker[0])/(x_max-stressMarker[0]))*(sysTime[-1]-im_time)+im_time

                    print(f"IM - RWY {points}")

            pointsarray[:,0] = pointsarray[:,0]-pointsarray[0,0] #Bring Base to 0

            pointsarray[:,0] = pointsarray[:,0]/pointsarray[-1,0] #Scale down to length 1

            

            # Load ECG Data 

            (gal_ecg,gc) = gen4file.readFolder(root)

            ecgruntime = len(gal_ecg)/fs_gal

            pointsarray[:,0] = pointsarray[:,0]*ecgruntime

            f = interpolate.interp1d(pointsarray[:,0], pointsarray[:,1])

            galtimeseries = np.linspace(0,ecgruntime,len(gal_ecg))

            new_pointsarray = f(galtimeseries)
            
            new_pointsarray = np.around(new_pointsarray,decimals=3)

            print(new_pointsarray.shape)
            
            print(len(gal_ecg))

            output_header = ['Time', 'Value', 'Stress']


            with open(split2[0]+'/P'+split2[1][-2]+split2[1][-1]+'Sc'+split1[1][-1]+'MRF.csv', 'w', newline=newline_char) as newfile:
                csvwriter = csv.writer(newfile, delimiter=delimiter_char)
                csvwriter.writerow(output_header)
                csvwriter.writerows(zip(galtimeseries, gal_ecg, new_pointsarray))

        print('<<---- Ende ---->>')
```
